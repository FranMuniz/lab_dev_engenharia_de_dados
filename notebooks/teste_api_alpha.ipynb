```python
import requests
import pandas as pd

API_KEY = "SUA_API_KEY_AQUI"
symbol = "AAPL"

url = f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={symbol}&outputsize=compact&apikey={API_KEY}"

response = requests.get(url)
data = response.json()

# Verificar se a requisição trouxe dados
if "Time Series (Daily)" in data:
    ts_data = data["Time Series (Daily)"]
    # Converter para DataFrame
    df = pd.DataFrame(ts_data).T  # Transpor: datas nas linhas
    df.index = pd.to_datetime(df.index)  # Converter index para datetime
    df = df.astype(float)  # converter valores para float
    print(df.head(50))
else:
    print("Erro ou limite de requisições:", data)
```
