```python
import boto3

s3 = boto3.resource(
    "s3",
    endpoint_url="http://localhost:9000",  # MinIO local
    aws_access_key_id="admin",
    aws_secret_access_key="franadmin",
    region_name="us-east-1"
)

buckets_para_deletar = [
    "bucket-teste-fran",
    "bucket-teste2-fran",
    "ingestao-teste-csv"
]

for bucket_name in buckets_para_deletar:
    bucket = s3.Bucket(bucket_name)
    print(f"Deletando bucket: {bucket_name}")

    # Deleta todos os objetos dentro do bucket
    bucket.objects.all().delete()

    # Deleta o bucket
    bucket.delete()
    print(f"Bucket {bucket_name} deletado com sucesso!\n")
```
